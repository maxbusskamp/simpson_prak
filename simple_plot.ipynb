```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
plt.rcParams['figure.dpi'] = 100

data_path = '/home/m/m_buss13/programming_part/2020/Day1/'
data = np.genfromtxt(data_path+'1000_16.xy', delimiter=' ')

fig = plt.figure(figsize=(6.5, 3), constrained_layout=True)
spec = gridspec.GridSpec(ncols=1, nrows=1, figure=fig)
f1_ax1 = fig.add_subplot(spec[0, 0])

f1_ax1.plot(data[:,0]/1000, data[:,1])
f1_ax1.set_xlabel(r'$\delta_\mathrm{iso}$ / kHz',
                      family='serif', fontsize=12)
```
