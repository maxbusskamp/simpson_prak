```python
#    PUT A UNIQUE TITLE HERE    #
#########Import Packages#########
import matplotlib.pyplot as plt #
import numpy as np              #
#################################

fig, ax =plt.subplots(figsize=(6.5, 4), facecolor='#f4f4f4', clear=True)  # Create a new Figure Object

# The following two lines can be duplicated to plot multiple datafiles. To generate Stackplots, change number added to dat[:,1]+0
dat = np.genfromtxt('./Day1/1000_16.xy', delimiter=' ')  # Read data from File into memory
plt.plot(dat[:,0]/1000, dat[:,1]+0, label='Label1')  # Plot data from memory and label uniquely.

# ax.set_xlim(-10.0, 10.0)  # Use this line if you want to show a special region
ax.set_xlabel(chr(948) + ' / kHz')  # X-Axis Label Text
ax.set_yticks([])  # Removes Y-Axis Label and Numbering, can also be left off
ax.autoscale(True, axis='y')  # Rescale Axis to fit all Plots

fig.savefig('./Day1/plot1.png',  # Filename for saved image
            format='png',  # Filetype, just leave at png
            dpi=600)  # Resolution, just leave at 600

plt.show()  # Show Figure in Notebook
```
